```python
# Initial imports
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path
```

```python
# Reading whale returns
df_whale = pd.read_csv('whale_returns.csv',index_col="Date",parse_dates=True, infer_datetime_format=True)
#count nulls
df_whale.isnull().sum()
# Drop nulls
df_whale = df_whale.dropna().copy()

df_whale.head()
```

```python
# Reading algorithmic returns
df_algo = pd.read_csv('algo_returns.csv',index_col="Date",parse_dates=True, infer_datetime_format=True)
#count nulls
df_algo.isnull().sum()
# Drop nulls
df_algo = df_algo.dropna().copy()

df_algo.head()
```

```python
# Reading S&P TSX 60 Closing Prices
df_sp_tsx = pd.read_csv('sp_tsx_history.csv',index_col="Date",parse_dates=True, infer_datetime_format=True)
# Check Data Types
df_sp_tsx.dtypes
# Fix Data Types
df_sp_tsx['Close'] = df_sp_tsx['Close'].str.replace("$","")
df_sp_tsx['Close'] = df_sp_tsx['Close'].str.replace(",","")
df_sp_tsx['Close'] = df_sp_tsx['Close'].astype('float')
```

```python
# Calculate Daily Returns and drop nulls
sp_tsx_daily_returns = df_sp_tsx.pct_change().dropna()
sp_tsx_daily_returns.head()
```

```python
# Rename `Close` Column to be specific to this portfolio.
sp_tsx_daily_returns = sp_tsx_daily_returns.rename(columns={
    "Close":"SP_TSX"
})
sp_tsx_daily_returns.head()
```

```python
# Join Whale Returns, Algorithmic Returns, and the S&P TSX 60 Returns into a single DataFrame with columns for each portfolio's returns.
joined_returns = pd.concat(
   [df_algo,sp_tsx_daily_returns,df_whale],axis='columns',join='inner'
)
joined_returns
```

```python
# Plot daily returns of all portfolios
joined_returns.plot()
```

```python
# Calculate cumulative returns of all portfolios
cumulative_returns = (1+joined_returns).cumprod()
# Plot cumulative returns
cumulative_returns.plot()
```

```python

# Calculate the daily standard deviations of all portfolios
# Calculate  the daily standard deviation of S&P TSX 60
daily_std_joined = joined_returns.std().sort_values(ascending=False)
daily_std_joined
```

```python

# Box plot to visually show risk
daily_std_joined.plot.box()
```

```python

# Determine which portfolios are riskier than the S&P TSX 60
print ("all else except 'PAULSON & CO.INC.' are risker than SP TSX 60")
```

```python
# Calculate the annualized standard deviation (252 trading days)
annualized_std = daily_std_joined * np.sqrt(252)
annualized_std
```

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window
# Plot the rolling standard deviation
joined_returns.rolling(window=21).std().plot()
```

```python
# Calculate the correlation
correltation = joined_returns.corr()
# Display de correlation matrix
correltation
```

```python
# Calculate covariance of a single portfolio
covariance_algo_1 = joined_returns['Algo 1'].rolling(window=21).cov(joined_returns['SP_TSX'])
covariance_algo_2 = joined_returns['Algo 2'].rolling(window=21).cov(joined_returns['SP_TSX'])
covariance_soros = joined_returns['SOROS FUND MANAGEMENT LLC'].rolling(window=21).cov(joined_returns['SP_TSX'])
covariance_pc = joined_returns['PAULSON & CO.INC. '].rolling(window=21).cov(joined_returns['SP_TSX'])
covariance_tiger = joined_returns['TIGER GLOBAL MANAGEMENT LLC'].rolling(window=21).cov(joined_returns['SP_TSX'])
covariance_berkshire = joined_returns['BERKSHIRE HATHAWAY INC'].rolling(window=21).cov(joined_returns['SP_TSX'])
# Calculate variance of S&P TSX
variance = joined_returns['SP_TSX'].rolling(window=21).var()
# Computing beta
algo_1_beta = covariance_algo_1 / variance
algo_2_beta = covariance_algo_2 / variance
soros_beta = covariance_soros / variance
pc_beta = covariance_pc / variance
tiger_beta = covariance_tiger / variance
berkshire_beta = covariance_berkshire / variance
# Plot beta trend
rolling_beta = pd.DataFrame ({
    "algo_1_beta":algo_1_beta,
    "algo_2_beta":algo_2_beta,
    "soros_beta":soros_beta,
    "pc_beta":pc_beta,
    "tiger_beta":tiger_beta,
    "berkshire_beta":berkshire_beta
})
rolling_beta.plot(figsize=(20,10))
```

```python
# Use `ewm` to calculate the rolling window
ewm_algo_1 = joined_returns['Algo 1'].ewm(halflife=21).mean()
ewm_algo_2 = joined_returns['Algo 2'].ewm(halflife=21).mean()
ewm_soros = joined_returns['SOROS FUND MANAGEMENT LLC'].ewm(halflife=21).mean()
ewm_pc = joined_returns['PAULSON & CO.INC. '].ewm(halflife=21).mean()
ewm_tiger = joined_returns['TIGER GLOBAL MANAGEMENT LLC'].ewm(halflife=21).mean()
ewm_berkshire = joined_returns['BERKSHIRE HATHAWAY INC'].ewm(halflife=21).mean()

ewm = pd.DataFrame ({
    "algo_1":ewm_algo_1,
    "algo_2":ewm_algo_2,
    "soros":ewm_soros,
    "pc":ewm_pc,
    "tiger":ewm_tiger,
    "berkshire":ewm_berkshire
})
ewm.plot(figsize=(20,10),title='exponentially moving average')
```

```python
# Annualized Sharpe Ratios
sharpe_ratios = (joined_returns.mean() * 252) / (joined_returns.std() * np.sqrt(252))
sharpe_ratios
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios.plot(kind="bar", title="Sharpe Ratios")
```

```python
print("BERKSHIRE HATHAWAY INC is outperforming than Algo 2, other than that overall algo trading is outporming than other portfolios/SP TSX")
```

```python
# Reading data from 1st stock
shop_data=Path("shop_historical.csv")
shop = pd.read_csv(shop_data,index_col='Date',infer_datetime_format=True, parse_dates=True).sort_index()
shop.head()
```

```python
# Reading data from 2nd stock
otex_data=Path("otex_historical.csv")
otex = pd.read_csv(otex_data,index_col='Date',infer_datetime_format=True, parse_dates=True).sort_index()
otex.head()
```

```python
# Reading data from 3rd stock
l_data=Path("l_historical.csv")
l = pd.read_csv(l_data,index_col='Date',infer_datetime_format=True, parse_dates=True).sort_index()
l.head()
```

```python
# Combine all stocks in a single DataFrame
joined_df = pd.concat([shop,otex,l],axis="rows",join="inner")
joined_df = joined_df.reset_index()
joined_df = joined_df.pivot_table(values="Close",index="Date",columns="Symbol")
joined_df.head()
```

```python
# Calculate daily returns
# Drop NAs
joined_daily_returns = joined_df.pct_change().dropna()

# Display sample data
joined_daily_returns.head()
```

```python
# Set weights
weights = [1/3, 1/3, 1/3]

# Calculate portfolio return
portfolio_returns = joined_daily_returns.dot(weights)
# Display sample data
portfolio_returns.head()
```

```python
# Join your returns DataFrame to the original returns DataFrame and drop NA
joined_returns_2 = pd.concat([joined_daily_returns,portfolio_returns],axis="columns",join="inner").dropna()
joined_returns_2.columns=['L','OTEX','SHOP','Portfolio']
                        
joined_returns_2.head()
```

```python
# Calculate the annualized `std`
daily_std = joined_returns_2.std().sort_values(ascending=False)
annualized_std = daily_std * np.sqrt(252)
annualized_std.sort_values(ascending=False)
```

```python
# Calculate rolling standard deviation
# Plot rolling standard deviation
joined_returns_2.rolling(window=21).std().plot()
```

```python
 # Calculate and plot the correlation
correlation = joined_returns_2.corr()
correlation
```

```python
import seaborn as sns
sns.heatmap(correlation, vmin=-1, vmax=1)
```

```python
# Calculate and plot Beta
all_returns = pd.concat([joined_returns_2,sp_tsx_daily_returns],axis="columns",join="inner").dropna()
rolling_covariance = all_returns['Portfolio'].rolling(window=60).cov(all_returns['SP_TSX'])
rolling_variance = all_returns['SP_TSX'].rolling(window=60).var()
beta = rolling_covariance / rolling_variance
beta.plot()
```

```python
# Calculate Annualzied Sharpe Ratios
sharpe_ratios = (all_returns.mean() * 252) / (all_returns.std() * np.sqrt(252))
sharpe_ratios
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios.plot(kind="bar")
```
